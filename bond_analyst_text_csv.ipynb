```python
import os
import re
```

```python
os.chdir("C:/Users/180309/mulcam_b/bond_report")
```

```python
path = os.getcwd()
file_list = os.listdir(path)
```

```python
text = []
date_text = []
for i in range(len(file_list)):
    f = open(file_list[i],'r',encoding='utf-8')
    text.append(f.readline())
    date_text.append('20'+re.findall('(.+?)_',file_list[0])[0])
```

```python
from datetime import datetime
date = []
for i in date_text:
    date.append(datetime.strptime(i,'%Y%m%d').strftime('%Y-%m-%d'))
```

```python
import pandas as pd
bond_analyst = pd.DataFrame([date,text],index=['date','text'])
bond_analyst = bond_analyst.T
```
