```python
import olefile

f = olefile.OleFileIO('금융통화위원회 의사록(2004년도 제25차)(2004.12.9).hwp')
encoded_text = f.openstream('PrvText').read()
decoded_text = encoded_text.decode('UTF-16')
print(decoded_text)
```
