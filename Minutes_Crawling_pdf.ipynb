```python
import requests
from bs4 import BeautifulSoup
from urllib import request

# 2006년 부터 2017년까지 의사록 pdf
for i in range(5, 31):
    url_prefix = 'http://www.bok.or.kr'
#     download_path = 'downloads\\'
    download_path = 'C:\\Users\\leeso\\TIL\\Project\\BOK_paper\\pdf min'
    url_page = '/portal/bbs/B0000245/list.do?menuNo=200761&pageIndex={}'.format(str(i))
#     print(url_prefix + url_page)
    resp = requests.get(url_prefix + url_page)

    soup = BeautifulSoup(resp.content, 'lxml')
    li_tag = soup.select('div.bdLine.type2 ul li')

    for li in li_tag:
        a_tag = li.select('div span a.secretSet')
        span_titlesub = li.select('div span a.secretSet span.titlesub')
        
        for span, a in zip(span_titlesub, a_tag):
            title = span.text
            link = a['href']
            file_name = title + '.pdf'
            w_url = url_prefix + link
#             print(w_url)
            
            d_resp = requests.get(w_url)
#             print(d_resp)
            d_soup = BeautifulSoup(d_resp.content, 'lxml')
            d_a_tag = d_soup.select('div.addfile ul li a')
#             print(d_a_tag[2])
            href = d_a_tag[2]['href'] #이 태그만 바뀜
            d_url = url_prefix + href
#             print(d_url)
            
            request.urlretrieve(d_url, download_path+'/'+file_name)
            print("download done : ", file_name)
```
